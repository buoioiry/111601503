保留分析Age、VIP、Destination、Homeplanet

剔除PassengerId、CryoSleep、Cabin、RoomService、FoodCourt、ShoppingMall、Spa、VRDeck、Name、Transported

```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#讀檔
df = pd.read_csv("train.csv")
df.head()
df.info()

#handle missing value看哪個欄位資料缺太多
df.isnull().sum() > len(df)/2 
len(df)/2

#填滿
df['Age'].fillna(df['Age'].value_counts().mean(),inplace=True)
df['VIP'].fillna(df['VIP'].value_counts().idxmax(),inplace=True)
df['Destination'].fillna(df['Destination'].value_counts().idxmax(),inplace=True)
df['HomePlanet'].fillna(df['HomePlanet'].value_counts().idxmax(),inplace=True)

#資料轉成01
df = pd.get_dummies(data=df, dtype = int, columns= ['HomePlanet','Destination','VIP'])
df.head()
df.isnull().sum()
df.head()

print(df.isnull().sum())
#只留Age VIP destination Homeplanet
df.drop(['PassengerId','CryoSleep','Cabin','RoomService','FoodCourt','ShoppingMall','VRDeck','Name','Spa'], axis=1, inplace=True)

X = df.drop(['Transported'],axis=1)
y = df['Transported']

df.info()
df.head()

#機器學習
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=67)


from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=200)
lr.fit(X_train,y_train)
predictions = lr.predict(X_test)


from sklearn.metrics import confusion_matrix,accuracy_score, recall_score, precision_score
accuracy_score(y_test,predictions)
recall_score(y_test,predictions)
precision_score(y_test,predictions)
pd.DataFrame(confusion_matrix(y_test,predictions),columns=['Predictnot Survived', 'PredictSurvived'],index=['Truenot Survived','TrueSurvived'])


# 模型匯出
import joblib
joblib.dump(lr,'Spaceship Titanic-LR-20231115.pkl',compress=3)
```
